```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=10,alpha=0.1) # RGB Hex Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',lw=1,alpha=0.5,ls='--',marker='o',markersize=20,markerfacecolor='yellow',markeredgewidth=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='violet',lw=2,ls='--',)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
