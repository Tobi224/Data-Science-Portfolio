```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My first plot in python that I understand')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
# first number is the left of the axes(left or right relative if it's two axis)
# second number is the bottom of the axes(up or down relative if it's two axis)
# third number is the width of the axes
# fourth number is the height of the axes
# all of them are in percentage
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Hi')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.6,0.2,0.2,0.2])

axes1.plot(x,y)
axes1.set_title('FIRST PLOT')

axes2.plot(y,x)
axes2.set_title('Second Plot')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for i in axes:
    i.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Hey')

axes[1].plot(y,x)
axes[1].set_title('1')
axes[1].set_ylabel(2)
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
plt.show()
```
