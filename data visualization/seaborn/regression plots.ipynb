```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(tips)
plt.show()
```

```python
sns.lmplot(y='tip',x='total_bill',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
plt.show()
# scatter_kws is for the size of a scatter plot
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill',y='tip',data=tips,row='time',col='day',hue='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
sns.set_style('darkgrid')
plt.show()
```
