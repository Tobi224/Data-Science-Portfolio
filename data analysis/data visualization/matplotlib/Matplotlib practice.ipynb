```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y,color='r',alpha=1)
ax2.plot(x,y,color='r')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

z = x**2
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

plt.show()
```

```python
fig, axes =plt.subplots(nrows=1,ncols=2)
plt.show()
```

```python
fig, axes =plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,ls='--',color='b',lw='3')
axes[1].plot(x,z,c='r',lw='3')
plt.show()
```

```python
fig, axes =plt.subplots(nrows=1,ncols=2,figsize=(14,2.2))

axes[0].plot(x,y,color='b',lw='5')
axes[1].plot(x,z,c='r',lw='3',ls='--')

axes[0].set_ylabel('z')
axes[0].set_xlabel('x')

axes[1].set_ylabel('y')
axes[1].set_xlabel('x')
plt.show()
```
