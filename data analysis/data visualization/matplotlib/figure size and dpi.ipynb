```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
import pandas as pd
```

```python
# figsize=(3,2) means the width and height of the plot in inches
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)

axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.set_title('Axes')
ax.set_ylabel('Amen')
ax.set_xlabel('Ok')
```
