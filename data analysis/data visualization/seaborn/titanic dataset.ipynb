```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python


sns.jointplot(x=titanic['fare'],y=titanic['age'])
plt.show()
```

```python

sns.distplot(titanic['fare'],kde=False,color='r',bins=30)
plt.show()
```

```python

sns.boxplot(x=titanic['class'],y=titanic['age'],palette='rainbow')
plt.show()
```

```python

sns.swarmplot(data=titanic,x='class',y='age',palette='Set2')
plt.show()
```

```python
titanic
```

```python

sns.countplot(x='sex',data=titanic,hue='sex')
plt.show()
```

```python

sns.heatmap(titanic.corr(numeric_only=True),cmap='coolwarm')
plt.title('titanic.corr()')
plt.show()
```

```python

sns.FacetGrid(data=titanic,col='sex').map(sns.distplot,'age',kde=False,bins=0)

plt.show()
```

```python
sns.FacetGrid(data=titanic,col='sex').map(plt.hist,'age',rwidth=5)

plt.show()
```
