```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
tips
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=False,dodge=True)
plt.show()
```

```python
# dodge is another name for split
sns.swarmplot(x='day',y='total_bill',data=tips,hue='day',dodge=True,palette='dark:b')
plt.show()
```

```python
# catplot is another name for factorplot
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',hue='day')
plt.show()
```
